```python
import os

# === OPTION 1: Hardcode your key here ===
GEMINI_API_KEY = "PASTE_YOUR_KEY_HERE"

# === OPTION 2: Use Colab secret (leave hardcode as-is to fall back) ===
if GEMINI_API_KEY == "PASTE_YOUR_KEY_HERE":
    try:
        from google.colab import userdata
        GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
    except Exception:
        GEMINI_API_KEY = os.environ.get('GEMINI_API_KEY', '')

os.environ['GEMINI_API_KEY'] = GEMINI_API_KEY
assert GEMINI_API_KEY, 'No API key set. Hardcode it or add a Colab secret named GEMINI_API_KEY.'
```

```python
from google import genai

client = genai.Client(api_key=os.environ.get('GEMINI_API_KEY'))

generation_config = {
    'max_output_tokens': 65536,
    'thinking_level': 'medium',
}

interaction = client.interactions.create(
    model='models/gemini-3.5-flash',
    input='Write an essay on Diwali 100 words',
    generation_config=generation_config,
)

print(interaction.output_text)
```
